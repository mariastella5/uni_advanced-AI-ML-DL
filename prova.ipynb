```python
import pandas as pd
import random
```

```python
dataset_origin = {'100k': 'u.data', '1M': 'ratings.dat'}

num_sample_data = '100k'
#DATA_PATH = 'drive/MyDrive/Colab Notebooks/2 - Data/{}'.format(dataset_origin[num_sample_data]) #change this with your directory 
DATA_PATH = 'data\MovieLens/{}'.format(dataset_origin[num_sample_data]) #change this with your directory 
```

```python
# load data
ml_100k = pd.read_csv(
	DATA_PATH, 
	sep="\t", 
	names = ['user_id', 'item_id', 'rating', 'timestamp'], 
	engine='python')
```

```python
ml_100k.head(2)#.groupby("user_id")['item_id'].count()
```

```python
interact_status = (ml_100k.groupby('user_id')['item_id']
                   .apply(set)
			        .reset_index()
			        .rename(columns={'item_id': 'interacted_items'}))
```

```python
interact_status
```

```python
item_pool = set(ml_100k['item_id'].unique())
num_ng_test = 100
```

```python
interact_status['negative_items'] = interact_status['interacted_items'].apply(lambda x: item_pool - x)
interact_status['negative_samples'] = interact_status['negative_items'].apply(lambda x: random.sample(sorted(x), num_ng_test))
```
